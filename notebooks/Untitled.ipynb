```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]
```

```python


# Sample dataset with three columns
dataset = np.array([
[4.37406,3.45067,0.0172741,0.0180721],
[4.39134,3.46874,0.0171074,0.0182301],
[4.40844,3.48697,0.0172691,0.0180771],
[4.42571,3.50505,0.0172379,0.0181068],
[4.44295,3.52316,0.017107,0.0182304],
[4.46006,3.54139,0.0172363,0.0181084],
[4.47729,3.5595,0.0170593,0.0182751]
])

# Separate columns into individual arrays
column1 = dataset[:, 0]
column2 = dataset[:, 1]
column3 = dataset[:, 2]
column4 = dataset[:, 3]

x = column1
y = column2
dx = column3
dy = column4
```

```python
plt.figure(figsize=(6, 4))  # Adjust the figure size if needed




for i in range(len(x)):
    plt.arrow(x[i], y[i], dx[i], dy[i], width=0.001, head_width=0.01, head_length=0.01, length_includes_head=True)
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Turned Arrows')
plt.xlim(0, 6)  # Adjust the x-axis limits if needed
plt.ylim(0, 12)  # Adjust the y-axis limits if needed
plt.grid(True)
plt.show()
```

```python
plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X-Y Graph')
plt.grid(True)
plt.show()
```
